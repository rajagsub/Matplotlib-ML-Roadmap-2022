### Analyzing Data from CSVs

```python
from collections import Counter #To create a counter for the items in the Languagesworkedwith 
```

```python
data = pd.read_csv('Datasets/data.csv')
data.head()
```

```python
lang_resposes = data['LanguagesWorkedWith']
lang_resposes
```

```python
language_counter = Counter()
```

```python
for response in lang_resposes:
    language_counter.update(response.split(';'))
```

```python
language_counter
```

```python
lang_most_common = language_counter.most_common(15)
```

```python
lang = []
popularity =[] 

for item in lang_most_common:
    lang.append(item[0])
    popularity.append(item[1])
```

```python
lang
```

```python
popularity
```

```python
lang.reverse()
popularity.reverse()
```

```python
plt.barh(lang,popularity)
plt.title('Most Popular Languages')
plt.xlabel('Number of people who uses')
plt.legend()
```

## Filling Area on Line Plots

```python
sample = pd.read_csv('Datasets/data_Sample.csv')
sample.head()
```

```python
ages = sample['Age']
dev_sal = sample['All_Devs']
py_sal = sample['Python']
js_sal = sample['JavaScript']
```

```python
plt.plot(ages,dev_sal,color='#eabd30', linestyle='--', label ='All_Devs')
plt.plot(ages,py_sal, color='#c23535', label = 'Python')


plt.title('Meadian Sal (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median SAL')
```

```python
overall_median = 57287
```

```python
plt.plot(ages,dev_sal,color='#eabd30', linestyle='--', label ='All_Devs')
plt.plot(ages,py_sal, color='#c23535', label = 'Python')

plt.fill_between(ages,py_sal,overall_median,
                 where=(py_sal>overall_median),
                 interpolate=True,
                 alpha=0.30)

plt.fill_between(ages,py_sal,overall_median,
                 where=(py_sal<=overall_median),
                 interpolate=True,
                 alpha=0.30)

plt.title('Meadian Sal (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median SAL')
```

```python
plt.plot(ages,dev_sal,color='#eabd30', linestyle='--', label ='All_Devs')
plt.plot(ages,py_sal, color='#c23535', label = 'Python')

plt.fill_between(ages,py_sal,dev_sal,
                 where=(py_sal>dev_sal),
                 interpolate=True,
                 alpha=0.30,label='Above Avg')

plt.fill_between(ages,py_sal,dev_sal,
                 where=(py_sal<=dev_sal),
                 interpolate=True,
                 alpha=0.30, label = 'Below Avg')

plt.title('Meadian Sal (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median SAL')
plt.legend()
```

## Plotting Time Series Data

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates,y,linestyle='solid')
plt.gcf().autofmt_xdate()
date_format = mpl_dates.DateFormatter('%b,%d,%y')
plt.gca().xaxis.set_major_formatter(date_format)
```

## Plotting Live Data in Real-Time

```python
import random
from itertools import count
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = [0, 1, 2, 3, 4, 5]
y_vals = [0, 1, 3, 2, 3, 5]

plt.plot(x_vals,y_vals)
```

```python
x = []
y = []
index = count()
```

```python
def animate(i):
    x.append(next(index))
    y.append(random.randint(0,5))
    
    plt.cla()
    plt.plot(x,y)
```

```python
ani = FuncAnimation(plt.gcf(),animate,interval=1000)
plt.tight_layout()
plt.show()
```
